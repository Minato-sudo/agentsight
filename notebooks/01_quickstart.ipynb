# AgentSight Quickstart
This notebook shows how to load and use the AgentSight model directly.

```
from product.agentsight_api import AgentSightDetector
detector = AgentSightDetector()
```

## Run Detection
Pass a trajectory into the detector.

```
query = 'Find flight to NYC'
trajectory = [
    {'step': 1, 'content': 'I will search for flights', 'tool_calls': [{'name': 'search', 'arguments': {}}], 'tool_responses': ['No flights found']}
]
result = detector.detect(query, trajectory)
print(result)
```
